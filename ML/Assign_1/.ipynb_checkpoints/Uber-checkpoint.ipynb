```python
df.dropna(inplace=True)
```

#### 3. Column pickup_datetime is in wrong format (Object). Convert it to DateTime Format

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime']).dt.strftime('%H').astype('float32')
```

```python
df.dtypes
```

```python
def distance(lat1, log1, lat2, log2):
    distance = np.sqrt((lat2 - lat1)**2 + (log2 - log1)**2)
    return distance

df['distance'] = distance(df['pickup_latitude'], df['pickup_longitude'], df['dropoff_latitude'], df['dropoff_longitude'])
```

```python
df.columns
```

#### 4. Outliers Removal

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.title("Before removing outliers")
plt.figure(figsize=(6,6))
sns.boxplot(df['fare_amount'])
plt.figure(figsize=(6,6))
sns.boxplot(df['distance'])

# remove outliers from pickup/dropoff locations
def remove_outliers(feature):
    global df
    q1 = np.percentile(df[feature], 25)
    q3 = np.percentile(df[feature], 75)
    IQR = q3 - q1

    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR

    df = df[ (df[feature] >= lower_bound) & (df[feature] <= upper_bound) ]

remove_outliers('fare_amount')
remove_outliers('distance')

plt.title("After removing outliers")
plt.figure(figsize=(6,6))
sns.boxplot(df['fare_amount'])
plt.figure(figsize=(6,6))
sns.boxplot(df['distance'])
```

#### 4. Correlation

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
```

#### 5. Models Implementation and Evaluation

```python
from sklearn.model_selection import train_test_split

X = df.drop('fare_amount', axis=1)
y = df['fare_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X_train, y_train)
prediction = regression.predict(X_test)
print(prediction)
```

```python
# Model Evaluation1
#1. R2 Score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, prediction)

#2.RMSE (Root Mean Squared Error)
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, prediction)
RMSE = np.sqrt(MSE)

print("R2 Score: ", r2)
print("Root Mean Squared Error: ",RMSE)
```

```python
#Random Forest Regression

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train, y_train)
prediction = rf.predict(X_test)
print(prediction)
```

```python
# Model Evaluation2
#1. R2 Score
r2 = r2_score(y_test, prediction)

#2.RMSE (Root Mean Squared Error)
MSE = mean_squared_error(y_test, prediction)
RMSE = np.sqrt(MSE)
RMSE

print("R2 Score: ", r2)
print("Root Mean Squared Error: ",RMSE)
```
