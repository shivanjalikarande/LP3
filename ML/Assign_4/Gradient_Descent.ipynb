### Gradient Descent Algorithm

##### Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2

```python
import matplotlib.pyplot as plt
```

```python
#function
def func(x):
    return (x+3)**2

#derivative
def grad(x):
    return 2*(x+3)
```

```python
#parameters for gradient descent
x = 2         #Initial Value
learning_rate = 0.1
epochs = 50   #No. of Iterations


#For plotting
x_vals = [x]
y_vals = [func(x)]
```

```python
#Gradient Descent Loop
tol = 1e-6
for i in range(epochs):
    x_new = x - learning_rate*grad(x)
    
    if abs(x_new - x) < tol:  # convergence check
        break

    x = x_new
    
    x_vals.append(x)
    y_vals.append(func(x))

    print(f'Iteration {i+1}: x = {x:.5f}    f(x)={func(x):.5f}')
```

```python
#Get Local Minimum
print(f"Local Minimum at : x = {x:.5f}  func(x) = {func(x):.5f}")
```

```python
#Visualisation - Gradient Descent Graph

x_plot = [i for i in range(-10,5)]
y_plot = [func(i) for i in x_plot]

plt.title("Gradient Descent to find local Minima")
plt.plot(x_plot,y_plot)
plt.scatter(x_vals,y_vals,color="red",label="Gradient Descent Steps")

plt.xlabel("func(x) = (x+3)^2")
plt.ylabel("y : grad(x)")

plt.grid()
```
